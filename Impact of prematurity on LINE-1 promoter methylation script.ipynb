# Consolidando Idade Gestacional

```python
dfWord = df.fillna(0)
IdadeGestacionalConsolidada = []
for index,row in dfWord.iterrows():
    if row['Idade gestacional (US) - semanas'] != 0:
      IdadeGestacionalConsolidada.append(row['Idade gestacional (US) - semanas'])
    elif row['Idade gestacional (Capurro) - semanas'] != 0:
      IdadeGestacionalConsolidada.append(row['Idade gestacional (Capurro) - semanas'])
    else:
      IdadeGestacionalConsolidada.append(row['Idade gestacional (DUM) - semanas'])
```

```python
df['IdadeGestacionalConsolidada'] = IdadeGestacionalConsolidada
```

# Definindo Rótulos

```python
maturidadeDoFeto = []
maturidadeDoFeto_agrupada = []
for index,row in df.iterrows():
  if row['IdadeGestacionalConsolidada'] >= 37:
    maturidadeDoFeto.append('Termos')
    maturidadeDoFeto_agrupada.append('Termos')
  elif row['IdadeGestacionalConsolidada'] < 37:
    maturidadeDoFeto_agrupada.append('Pré-termos')
    if row['IdadeGestacionalConsolidada'] < 28:
      maturidadeDoFeto.append('Prematuro extremo')
    elif row['IdadeGestacionalConsolidada'] >= 28:
      maturidadeDoFeto.append('Muito prematuro')
```

```python
def compararDescricoes(colname:str):
  d1 = df[[colname,'maturidadeDoFeto_agrupada']].loc[df['maturidadeDoFeto_agrupada'] == 'Termos'].describe(include='all')
  d2 = df[[colname,'maturidadeDoFeto_agrupada']].loc[df['maturidadeDoFeto_agrupada'] != 'Termos'].describe(include='all')
  return pd.DataFrame(zip(d1[colname],d2[colname]),columns=['Termos','Pré-termos'],index=d1.index)

descricaoIdadeMaterna = compararDescricoes('Idade materna')
descricaoIdadeMaterna
```

```python
def compararMedianas(targetCol):
  return {
      "Termos":round(df[targetCol].loc[df['maturidadeDoFeto_agrupada'] == 'Termos'].median(),1),
      "Pré-termos":round(df[targetCol].loc[df['maturidadeDoFeto_agrupada'] != 'Termos'].median(),1)}

compararMedianas('Idade materna')
```

```python
def compararIQR(targetCol):
  describeAtermos = df[targetCol].loc[df['maturidadeDoFeto_agrupada']=='Termos'].describe()
  describePretermos =  df[targetCol].loc[df['maturidadeDoFeto_agrupada']!='Termos'].describe()
  return {
      'IQR Termos':f"{round(describeAtermos['25%'],1)} - {round(describeAtermos['75%'],1)}",
      'IQR Pré-termos':f"{round(describePretermos['25%'],1)} - {round(describePretermos['75%'],1)}"
  }

compararIQR('Idade materna')
```

```python
def showPercentils(targetCol):
  valuesCount = df[[targetCol,'maturidadeDoFeto_agrupada']].value_counts()
  percentil = []
  for row in range(0,len(valuesCount.values)):
    key = valuesCount.index[row][1]
    n_percentil = valuesCount.values[row]/nsamples.get(key)
    percentil.append(round(n_percentil*100,1))
  countPercentil = pd.DataFrame(zip(valuesCount.values,percentil),columns=['n','%'],index = valuesCount.index)
  return countPercentil

showPercentils('Escolaridade materna')
```

```python
def ploPairtViolin(x,y,title=None,ylabel=None,xlabel=None,lims=None,annotation=True):
  # Configure Plot
  plotParams = {
      'data':df,
      'x':x,'y':y,
      'palette': (sns.color_palette("Paired")[0],sns.color_palette("Paired")[4],sns.color_palette("Paired")[6])
      }

  ax = sns.violinplot(**plotParams)

  # loc values
  pairs = [('Termos', 'Pré-termos')]
  loc_class1 = df[y].loc[df[x] == 'Termos']
  loc_class2 = df[y].loc[df[x] != 'Termos']
  loc_values = [loc_class1,loc_class2]
  nvalues = [len(loc_class1.dropna().to_list()),len(loc_class2.dropna().to_list())]
  if annotation:
    # Configura Annotator
    annotator = Annotator(ax=ax,pairs=pairs,**plotParams)
    annotator.configure(loc='inside',verbose=False)
    # Anotate Medians
    median = [loc.median() if (loc.dropna().values.shape[0])>3 else 'NaN' for loc in loc_values ]
    for index in range(len(median)):
      if type(median[index]) != str:
        median_value = round(median[index],2)
        ax.annotate(f' {median_value}', xy=(index,median_value),horizontalalignment='left')
        ax.annotate(f'N = {nvalues[index]}', xy=(index,lims[0]+2))

    # Anotate P Values by MW
    try:
      pvalue = stats.mannwhitneyu(
            loc_class1.to_list(),
            loc_class2.to_list(),
            alternative='two-sided',nan_policy='omit').pvalue
      value = round(pvalue,3)
      if value == 0:
        formatted_pvalues = [f'p<0.001']
      else:
        formatted_pvalues = [f'p={value}']
      annotator.set_custom_annotations(formatted_pvalues)
      annotator.annotate()
    except ValueError: # To avoid zero-values on axis
      pass

  # Annotate Title
  if title is None:
    sitio_CPG = y[y.find('cpg')+3:y.find('cpg')+4]
    dia = y[y.find('d')+1]
    ax.set_title(f'Resultado Seq1 CpG{sitio_CPG} D{dia}')
  else:
    ax.set_title(title)

  # Configure limits and labels
  ax.set_ylim(lims)
  ax.set_ylabel(ylabel)
  ax.set_xlabel(xlabel)
  plt.show()

ploPairtViolin(
    x = 'maturidadeDoFeto_agrupada',
    y = 'res_seq1_cpg1_d1_IDOR',
    ylabel = 'Metilação LINE-1 (%)',
    lims = (50,80)
)
```

```python
def plotTripleViolin(x,y,title=None,ylabel=None,xlabel=None,lims=None,annotation=True):
  # Configure Plot
  plotParams = {
      'data':df,
      'x':x,'y':y,
      'palette': (sns.color_palette("Paired")[0],sns.color_palette("Paired")[4],sns.color_palette("Paired")[6])
      }

  ax = sns.violinplot(**plotParams)
  line_height = 0.08
  # loc values
  pairs = [('Termos', 'Muito prematuro'),('Termos', 'Prematuro extremo'),('Muito prematuro', 'Prematuro extremo')]
  loc_class1 = df[y].loc[df[x] == 'Termos']
  loc_class2 = df[y].loc[df[x] == 'Muito prematuro']
  loc_class3 = df[y].loc[df[x] == 'Prematuro extremo']
  loc_values = [loc_class1,loc_class2,loc_class3]
  nvalues = [loc_class1.dropna().to_list(),loc_class2.dropna().to_list(),loc_class3.dropna().to_list()]
  if annotation:
    # Configura Annotator
    annotator = Annotator(ax=ax,pairs=pairs,**plotParams)
    annotator.configure(loc='inside',line_height=line_height)

    # Anotate Medians
    median = [loc.median() if (loc.dropna().values.shape[0])>3 else 'NaN' for loc in loc_values ]
    for index in range(len(median)):
      if type(median[index]) != str:
        median_value = round(median[index],2)
        ax.annotate(f' {median_value}', xy=(index,median_value),horizontalalignment='left')
        ax.annotate(f'N = {len(nvalues[index])}', xy=(index,lims[0]+2))
    # Anotate P Values by MW
    try:
      p_values = [
          stats.mannwhitneyu(
            df[y].loc[df[x] == 'Termos'].to_list(),
            df[y].loc[df[x] == 'Muito prematuro'].to_list(),
            alternative='two-sided',nan_policy='omit').pvalue,
          stats.mannwhitneyu(
            df[y].loc[df[x] == 'Termos'].to_list(),
            df[y].loc[df[x] == 'Prematuro extremo'].to_list(),
            alternative='two-sided',nan_policy='omit').pvalue,
          stats.mannwhitneyu(
            df[y].loc[df[x] == 'Muito prematuro'].to_list(),
            df[y].loc[df[x] == 'Prematuro extremo'].to_list(),
            alternative='two-sided',nan_policy='omit').pvalue]

      formatted_pvalues = []
      for p in p_values:
        value = round(p,3)
        if value == 0:
          formatted_pvalues.append(f'p<0,001')
        else:
          formatted_pvalues.append(f'p={value}')
      annotator.set_custom_annotations(formatted_pvalues)
      annotator.annotate()

    except : # To avoid zero-values on axis
      pairs = [('Muito prematuro', 'Prematuro extremo')]
      annotator = Annotator(ax=ax,pairs=pairs,**plotParams)
      annotator.configure(loc='inside',line_height=line_height)
      p_values = [
          stats.mannwhitneyu(
            df[y].loc[df[x] == 'Muito prematuro'].to_list(),
            df[y].loc[df[x] == 'Prematuro extremo'].to_list(),
            alternative='two-sided',nan_policy='omit').pvalue]

      formatted_pvalues = []
      for p in p_values:
        value = round(p,3)
        if value == 0:
          formatted_pvalues.append(f'p<0,001')
        else:
          formatted_pvalues.append(f'p={value}')
      annotator.set_custom_annotations(formatted_pvalues)
      annotator.annotate()
  # Annotate Title
  if title is None:
    sitio_CPG = y[y.find('cpg')+3:y.find('cpg')+4]
    dia = y[y.find('d')+1]
    ax.set_title(f'Resultado Seq1 CpG{sitio_CPG} D{dia}')
  else:
    ax.set_title(title)

  # Configure limits and labels
  ax.set_ylim(lims)
  ax.set_ylabel(ylabel)
  ax.set_xlabel(xlabel)
  plt.show()

for i in variaveisMetilacao['idor'].values():
  for col in i:
    plotTripleViolin(
          x = 'maturidadeDoFeto',
          y = col,
          ylabel = 'Metilação LINE-1 (%)',
          lims = (50,100))
```

# Evolução temporal e ajuste linear das medianas de metilação

```python
from scipy.stats import linregress
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def compararMedianas(series):
    mediana = series.median()
    return {'Prematuros': mediana}

def compararIQR(series):
    q25, q75 = series.quantile([0.25, 0.75])
    return {'IQR Prematuros': f"{q25}-{q75}"}

def plotLineMedianErr(columnSets, media=False):
    indexD = ['D0', 'D5', 'D30', 'D90']
    if not media:
        cpg_N = int(columnSets[0][columnSets[0].find('cpg') + 3]) + 20  # Adiciona 20 à numeração do CpG
    else:
        cpg_N = "21-24"

    # Grupos específicos e agrupados
    medianasLine = {
        'Prematuros, em geral': [],
        'Muito prematuros': [],
        'Prematuros extremos': []
    }

    iqrLine = {
        'Prematuros, em geral': [],
        'Muito prematuros': [],
        'Prematuros extremos': []
    }

    for timepoint in columnSets:
        # Dados agrupados para 'Prematuros, em geral'
        pre_termo_data = df[df['maturidadeDoFeto'] != 'Termos'][timepoint].dropna()
        if not pre_termo_data.empty:
            medianasLine['Prematuros, em geral'].append(compararMedianas(pre_termo_data))
            iqrLine['Prematuros, em geral'].append(compararIQR(pre_termo_data))
        else:
            medianasLine['Prematuros, em geral'].append({'Prematuros': np.nan})
            iqrLine['Prematuros, em geral'].append({'IQR Prematuros': 'NaN-NaN'})

        for group in ['Muito prematuro', 'Prematuro extremo']:
            group_label = 'Muito prematuros' if group == 'Muito prematuro' else 'Prematuros extremos'
            data = df[df['maturidadeDoFeto'] == group][timepoint].dropna()
            if not data.empty:
                medianasLine[group_label].append(compararMedianas(data))
                iqrLine[group_label].append(compararIQR(data))
            else:
                medianasLine[group_label].append({'Prematuros': np.nan})
                iqrLine[group_label].append({'IQR Prematuros': 'NaN-NaN'})

    plt.figure(figsize=(12, 8))

    colors = {'Prematuros, em geral': '#AEC6CF', 'Muito prematuros': '#77DD77', 'Prematuros extremos': '#FFB347'}
    markers = {'Prematuros, em geral': 'o', 'Muito prematuros': 's', 'Prematuros extremos': 'D'}

    x_offset = {'Prematuros, em geral': -0.2, 'Muito prematuros': 0, 'Prematuros extremos': 0.2}

    for group in medianasLine.keys():
        med = [float(i['Prematuros']) for i in medianasLine[group] if not pd.isna(i['Prematuros'])]
        iqr_25 = [float(iqr['IQR Prematuros'].split('-')[0]) for iqr in iqrLine[group] if 'NaN' not in iqr['IQR Prematuros']]
        iqr_75 = [float(iqr['IQR Prematuros'].split('-')[1]) for iqr in iqrLine[group] if 'NaN' not in iqr['IQR Prematuros']]

        if len(med) == len(indexD) and len(iqr_25) == len(indexD) and len(iqr_75) == len(indexD):
            medianaDF = pd.DataFrame(zip(med, iqr_25, iqr_75), index=indexD, columns=[f'Metilação CpG{cpg_N} LINE-1 (%)', '25%', '75%'])

            sns.scatterplot(data=medianaDF, x=[i + x_offset[group] for i in range(len(medianaDF))], y=f'Metilação CpG{cpg_N} LINE-1 (%)', color=colors[group], marker=markers[group], s=100, label=group)

            for i, row in enumerate(medianaDF.iterrows()):
                plt.errorbar(
                    x=i + x_offset[group],
                    y=row[1][f'Metilação CpG{cpg_N} LINE-1 (%)'],
                    yerr=[[row[1][f'Metilação CpG{cpg_N} LINE-1 (%)'] - row[1]['25%']], [row[1]['75%'] - row[1][f'Metilação CpG{cpg_N} LINE-1 (%)']]],
                    color=colors[group],
                    capsize=5,
                    fmt='o',
                    elinewidth=2,
                    markeredgewidth=2
                )

            # Fit linear regression
            if len(medianaDF) > 1:
                slope, intercept, r_value, p_value, std_err = linregress(range(1, len(medianaDF) + 1), medianaDF[f'Metilação CpG{cpg_N} LINE-1 (%)'])
                # Plot linear regression line
                x_values = np.arange(1, len(medianaDF) + 1)
                y_values = slope * x_values + intercept
                plt.plot([i + x_offset[group] for i in range(len(indexD))], y_values, color=colors[group], label=f'{group} (R²={r_value**2:.2f}, Slope = {round(slope,2)}, p-value = {round(p_value,2)})')

    plt.xlabel('Tempo de vida (dias)')
    plt.xticks(ticks=np.arange(len(indexD)), labels=indexD)  # Define as marcações do eixo X
    if not media:
        plt.ylabel(f'Metilação CpG{cpg_N} LINE-1 (%)')
        plt.title(f'Evolução Temporal da metilação no CpG{cpg_N} de LINE-1')
    else:
        plt.ylabel(f'Metilação CpG(21-24) de LINE-1 (%)')
        plt.title(f'Evolução temporal da metilação nos CpG(21-24) de LINE-1')

    plt.legend()
    plt.show()
```
